```python
import pandas as pd
import numpy as np

df = pd.read_csv('diabetes/diabetes.csv')

# View structure
print("Shape:", df.shape)
print(df.head())
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt

df.hist(bins=20, figsize=(14, 10), edgecolor='black')
plt.suptitle("Histograms of Features")
plt.tight_layout()
plt.show()
```

```python
df.groupby('Outcome').mean().plot(kind='bar', figsize=(12, 6))
plt.title("Mean Feature Values Grouped by Diabetes Outcome")
plt.ylabel("Mean Value")
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

sns.scatterplot(data=df, x='Glucose', y='BMI', hue='Outcome', palette='Set1')
plt.title("Glucose vs BMI by Diabetes Outcome")
plt.grid(True)
plt.show()

sns.scatterplot(data=df, x='Glucose', y='Age', hue='Outcome', palette='Set1')
plt.title("Glucose vs Age by Diabetes Outcome")
plt.grid(True)
plt.show()

```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.boxplot(x='Outcome', y='BMI', data=df)
plt.title("BMI Distribution by Diabetes Outcome")
plt.grid(True)
plt.show()

sns.boxplot(x='Outcome', y='Glucose', data=df)
plt.title("Glucose Distribution by Diabetes Outcome")
plt.grid(True)
plt.show()
```
